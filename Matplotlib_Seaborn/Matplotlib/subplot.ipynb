```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(2,3,1) #Aqui se define el numero de filas, el numero de columnas y la posicion en la que voy a insertar mi grafico, que en este caso es la primera
plt.plot(x,y)

plt.subplot(2,3,2) #Aqui se inserta el segundo grafico en la segunda posicion.
plt.hist(y)

plt.show()
```
