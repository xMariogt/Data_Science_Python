```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, 'b-')
ax.plot(x,x+3, 'y.-')
ax.plot(x,x+4, 'go:')
```

USANDO PYPLOT

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, color= 'green', alpha=0.8, linewidth= 8) #Alpha da transparencia
ax.plot(x,x+2, color= 'blue', linestyle= '--', marker= 'x')
ax.plot(x,x+3, color= '#66FF66', linestyle= 'dashed', marker= 'P', markersize=8, markerfacecolor= 'white')
ax.plot(x,x+4, color= '#CCFFCC', linestyle= ':')
```
