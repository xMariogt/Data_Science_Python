```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50, 100)
```

## HISTOGRAMA

```python
plt.hist(data, bins=20) #bins devuelve la frecuencia de las barras
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step') 
plt.show()
```

## BOXPLOT

```python
data = np.append(data, 200) #este es un outlayer, un dato fuera de rango
plt.boxplot(data,vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Scatter o grafico de dispersion

```python
N = 50
x = np.random.rand(50)
y = np.random.rand(50)
area = (30*np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area, c=colors, marker='o', alpha=0.5) #s y c dan lugar a analizar dos variables mas dentro del mismo scatter 
#y alpha es el grado de transparencia de los objetos a visualizar
plt.show()
```
