```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
fig.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #Crea un tipo de arreglo de axes donde se define la cantidad de filas y columnas.
axes[0].plot(x,y, 'b')
axes[1].plot(y,x, 'r')
```

```python
fig, axes = plt.subplots(2, 4)

axes[0,0].plot(x,np.cos(x), 'y')
axes[0,1].plot(x,np.sin(x), 'r')
axes[0,2].plot(x,np.tan(x), 'b')
axes[0,3].plot(x,np.cos(x)**2, 'g')
axes[1,0].plot(y,x, 'y')
axes[1,1].plot(y,y**2, 'r')
axes[1,2].plot(y,np.tan(y), 'b')
axes[1,3].plot(y,x**2, 'g')
fig.tight_layout()
```
