```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex') #hue segmenta la informacion
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')#kind especifica el tipo de grafico
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', legend=True, palette='dark') #legend quita o pone las leyendas
plt.show()
```
