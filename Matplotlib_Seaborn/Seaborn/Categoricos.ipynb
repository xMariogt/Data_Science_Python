```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data=tips,y='day', hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True, color='0')
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True, palette='dark', split=True)
plt.show()
```

```python
sns.catplot(data=tips,x='day', y='total_bill',hue='sex', dodge=True, kind='swarm') #Permite usar todas las anteriores solo modificando kind
plt.show()
```
