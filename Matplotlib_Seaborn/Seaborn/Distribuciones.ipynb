```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips, x='tip', bins=15) #Bins hace referencia al numero de barras a utilizar 
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True) #Cumulative va acumulando las frecuencias en el grafico
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex') 
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex',stat='frequency') 
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex',stat='frequency',multiple='dodge') #multiple indica la forma en la que queremos visualizar las barras
plt.show()
```

DIAGRAMA DE DENSIDAD KDE

```python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='dodge')#Usar este es mejor ya que incluye todos los anteriores.
plt.show()
```
