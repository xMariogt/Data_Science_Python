```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100
x = np.linspace(-10, 10, N)
```

# Lineales

```python

m = -2
b = 5


def f(x):
    return m*x+b
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
plt.show()
```

# Polinomicas

```python
def f(x):
    return x**3
```

```python
y = f(x)

plt.plot(x, y)
plt.grid()
plt.show()
```

# Trascendentes

## Trigonometricas

```python
def f(x):
    return np.cos(x)


y = f(x)

plt.plot(x, y)
plt.show()
```

```python
def f(x):
    return np.sin(x)


y = f(x)

plt.plot(x, y)
plt.show()
```

```python
def f(x):
    return np.tan(x)


y = f(x)

plt.plot(x, y)
plt.show()
```

## Exponencial

```python
a = 4


def f(x):
    return a**x


y = f(x)

plt.plot(x, y)
plt.show()
```

```python

def f(x):
    return np.exp(x)


y = f(x)

plt.plot(x, y)
plt.show()
```

## Logaritmo

```python
x = np.linspace(0.1, 256, 1000)


def f(x):
    return np.log2(x)


y = f(x)

plt.plot(x, y)
plt.show()
```

# Funciones seccionadas

```python
x = np.linspace(-10, 10, N)

def H(x):
    Y = np.zeros(len(x))
    for idx,x in enumerate(x):
        if x >= 0:
            Y[idx]=1
    return Y    

plt.plot(x,H(x))
plt.show()
```

# Funciones compuestas

```python
def g(x):
    return x**2

def f(x):
    return np.sin(x)

y = f(g(x))

plt.plot(x,y)
plt.show()
```
